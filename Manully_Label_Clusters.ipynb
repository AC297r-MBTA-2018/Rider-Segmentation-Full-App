```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hier = 'hierarchical'
algo = 'lda'
profile_filename = 'data/cached_profiles/2017-Nov/{}_cluster_profiles_1711_1_0_{}.csv'.format(hier, algo)
df = pd.read_csv(profile_filename, index_col=0)
print(df.shape)
hr_cols = [col for col in df.columns if 'hr_' in col]

X_1D = df[hr_cols].values
X = X_1D.reshape((X_1D.shape[0], 7, 24))

for i in range(X_1D.shape[0]):
    plt.imshow(X[i])
    plt.show()
```

```python
labels = [2,4,5,6,6,2,4,2,4]
df['manual_label'] = labels
df
```

```python
df.to_csv(profile_filename)
```
